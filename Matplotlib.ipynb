```python
t = np.arange(0., 5., 0.2) 
s1=t**2 #niligris
s2=t**3 #more
s3=t**4 #reliance
#blue dashes, red squares and green triangles
plt.plot(t, s1, 'b--', label='niligris')#   'rs',   'g^')   # b-- means blue with dash line 
plt.legend() # add legend based on line labels
plt.show() # ending first graph
plt.plot(t,s2, 'rs', label='More')# rs means red with sqaures 
plt.legend() # add legend based on line labels
plt.show()
plt.plot(t,s3, 'g^', label='Reliance')# g^ means green triangle 
plt.legend() # add legend based on line labels
plt.xlabel('months')
plt.ylabel('sales in lakhs')
plt.show()
```

### Bar graph

```python
a=[1,4,10]
b=[12,2,6]
c=[6,8,1]
d=[1,5,7]
plt.bar(a,b,align='center', color ='b') # bar alignment to center
plt.title('Results')
plt.grid()
plt.xlabel('time in hrs')
plt.ylabel('journey in KMs')
plt.show()
print ('\n\n\n')
plt.bar(c,d, align='edge', color = 'orange')
plt.grid()
plt.show()
```

### Pie Charts 

```python
desig ='Softwares', 'Scientists','Engineers', 'Managers', 'testers', 'dev'
sizes =[100,130,65,25, 20, 30] 
c=['blue', 'green', 'yellow','red', 'orange', 'black']  
plt.pie(sizes,labels=desig, colors=c)
plt.axis('equal')
plt.show()
```

```python
desig ='Softwares', 'Scientists','Engineers', 'Managers','testers', 'dev'
sizes =[100,130,65,25, 20, 30]
colors=['blue', 'green', 'yellow','red', 'orange', 'black']
  
plt.pie(sizes,labels=desig,colors=colors,autopct='%1.2f%%' ) #1.1 - 1 decimal, 1.2 - 2 decimal ., 1- roundoff...
plt.axis('equal')
plt.show()
```

```python
desig ='Softwares', 'Scientists','Engineers', 'Managers','testers', 'dev'
sizes =[100,130,65,25, 20, 30]
colors=['blue', 'green', 'yellow','red', 'orange', 'black']
distance=(0.1,0,0,0.2,0,0) # distance
plt.pie(sizes,labels=desig,colors= colors,autopct='%1.0f%%', explode=distance)
plt.axis('equal')
plt.show()
```
